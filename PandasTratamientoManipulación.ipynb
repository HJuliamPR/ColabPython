```python
import pandas as pd
```

```python
DataHosting = pd.read_json('/content/drive/MyDrive/GoogleColab/datos_hosting.json')
DataInmuebles = pd.read_json('/content/drive/MyDrive/GoogleColab/inmuebles_disponibles.json')
```

```python
DataHosting = pd.json_normalize( DataHosting['info_inmuebles'] )
DataHosting.head()
```

```python
DataHosting.info()
```

## Datos Numéricos



```python
ColumnsHosting = list( DataHosting.columns )
ColumnsHosting
```

```python
# Lo que hace es desagregar los valores tipo lista y crear más registros
DataHosting = DataHosting.explode(ColumnsHosting[3:])
```

```python
# Lo que hace es volver a contar el índice y aplicar directamente en el DataFrame
DataHosting.reset_index(inplace = True, drop = True)
DataHosting.head(3)
```

```python
DataHosting.info()
```

```python
# Biblioteca para trabajar con datos numéricos Aplicar fórmulas matemáticas.
import numpy as np
```

```python
# Se le asigna el tipo de variable de la columna a entero
DataHosting['max_hospedes'] = DataHosting['max_hospedes'].astype(np.int64)
```

```python
ColumnINT64 = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
DataHosting[ColumnINT64] = DataHosting[ColumnINT64].astype(np.int64)
```

```python
# Se le asigna el tipo de variable de la columna a flotante
DataHosting['evaluacion_general'] = DataHosting['evaluacion_general'].astype(np.float64)
```

```python
# Método para transformar los datos en la columna indicada.
# Con la función homónima lambda cada uno de los parámetros.
# el método apply solo funciona para una única columna o sea, una serie.
DataHosting['precio'] = DataHosting['precio'].apply(lambda x: x.replace('$','').replace(',','').strip()).astype(np.float64)
```

```python
DataLocacao = pd.read_json('/content/drive/MyDrive/GoogleColab/dados_locacao_imoveis.json')
DataVentas = pd.read_json('/content/drive/MyDrive/GoogleColab/dados_vendas_clientes.json')
```

```python
DataLocacao = pd.json_normalize(DataLocacao['dados_locacao'])
DataLocacao.head(3)
```

```python
DataVentas = pd.json_normalize( DataVentas['dados_vendas'] )
DataVentas.head(3)
```

```python
DataLocacao = DataLocacao.explode( ['datas_combinadas_pagamento',	'datas_de_pagamento',	'valor_aluguel'] )
DataLocacao.reset_index(inplace = True, drop = True)

DataVentas = DataVentas.explode( ['Cliente', 'Valor da compra'] )
DataLocacao.reset_index(inplace = True, drop = True)
```

```python
DataLocacao['valor_aluguel'] = DataLocacao['valor_aluguel'].apply( lambda x: x.replace('$','').replace('reais','').replace(',','.').strip() ).astype(np.float16)

DataVentas['Valor da compra'] = DataVentas['Valor da compra'].apply( lambda x: x.replace('R','').replace(',','.').strip() ).astype(np.float16)
```
