# Leyendo SQL

```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
# DataClientsBank = pd.read_csv('/content/drive/MyDrive/GoogleColab/clientes_banco.csv')
# SQLite, MySQL, PostgreSQL, Oracle, MongoDB.
SQL_Engine = create_engine('sqlite:///:memory:')
```

```python
DataClientsBank = pd.read_csv('/content/drive/MyDrive/GoogleColab/clientes_banco.csv')
```

```python
DataClientsBank.head()
```

```python
DataClientsBank.to_sql('Clients', SQL_Engine, index=False)
```

```python
inspector = inspect(SQL_Engine)
inspector.get_table_names()
```

## Haciendo una consulta Query

```python
Query = 'SELECT * FROM Clients WHERE Categoria_de_renta = "Empleado"'
```

```python
# This text method is imported from AQL alchemy
Employees = pd.read_sql( sql=text(Query), con = SQL_Engine.connect() )
Employees
```

```python
Employees.to_sql('Employees', SQL_Engine, index=False)
```

```python
pd.read_sql_table("Employees", con = SQL_Engine.connect())
```

```python
pd.read_sql_table("Employees", con = SQL_Engine.connect(), columns=['ID_Cliente',	'Edad',	'Grado_estudio',	'Estado_civil'])
```

## Actualizando un banco de datos

```python
Query_1 = 'SELECT * FROM Clients'
```

```python
DataClients = pd.read_sql( sql = text(Query_1), con = SQL_Engine.connect())
```

```python
DataClients.columns
```

```python
# Delete rows
from sqlalchemy.exc import SQLAlchemyError
Query_2 = 'DELETE FROM Clients WHERE ID_Cliente=5008804'
try:
  #Return the rows where the query its been applied
  r_set = SQL_Engine.connect().execute( text(Query_2) )
except SQLAlchemyError as error:
  print(error)
else:
  print('Rows deleted: ', r_set.rowcount)
```

```python
# Update rows
Query_2 = 'UPDATE Clients SET Grado_estudio = "Nivel superior" WHERE ID_Cliente=5008808'
try:
  #Return the rows where the query its been applied
  r_set = SQL_Engine.connect().execute( text(Query_2) )
except SQLAlchemyError as error:
  print(error)
else:
  print('Rows updated: ', r_set.rowcount)
```
