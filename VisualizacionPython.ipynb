# Data Visualization


```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-es-cursos/2063-visualizacion-de-datos-con-bibliotecas-de-python/main/inmigrantes_canada.csv'
DataInmigrant = pd.read_csv(url)
display(DataInmigrant.head())
```

```python
DataInmigrant.info()
```

```python
# Hace que el índice sean los países.
DataInmigrant.set_index('Pais', inplace = True)
```

```python
DataInmigrant.head()
```

```python
Years = list(map(str,range(1980, 2014)))
# dataframe.loc[row_label, column_label]
Colombia = DataInmigrant.loc['Colombia', Years]
```

```python
ColombiaDictionary = {'Year':Colombia.index.tolist(),
                      'NumberMigrants':Colombia.values.tolist()}
```

```python
DataFrameColombia = pd.DataFrame(ColombiaDictionary)
DataFrameColombia.head()
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[8,4])
plt.plot(DataFrameColombia['Year'], DataFrameColombia['NumberMigrants'])
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010" ])
plt.title("Colmbian Migration to Canda")
plt.xlabel("Year")
plt.ylabel("Number of Migrants")
plt.show()
```

```python
#  Figura y gráfica
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(DataFrameColombia['Year'], DataFrameColombia['NumberMigrants'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Colombian Migration\n 1980 - 2013')
ax.set_xlabel('Year')
ax.set_ylabel("Number of Migrants")
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize = (10,3) )

axs[0].plot( DataFrameColombia['Year'], DataFrameColombia['NumberMigrants'] )
axs[0].set_title('Colombian Migration to Canada\n1980 - 2013')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Number of migrants')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot( DataFrameColombia['NumberMigrants'] )
axs[1].set_title('Boxplot Colombian Migration to Canada\n1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Number of migrants')
axs[1].grid()
plt.show()
```

```python
DataFrameColombia.describe()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Migration from biggest Latin american Countries')

axs[0,0].plot(DataInmigrant.loc['Colombia', Years])
axs[0,0]. set_title('Colombia')

axs[0,1].plot(DataInmigrant.loc['Brasil', Years])
axs[0,1]. set_title('Brasil')

axs[1,0].plot(DataInmigrant.loc['Argentina', Years])
axs[1,0]. set_title('Argentina')

axs[1,1].plot(DataInmigrant.loc['Perú', Years])
axs[1,1]. set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Year')
  ax.set_ylabel('Number of migrants')
  ax.grid()

plt.show()
```

# Desafio

```python

# dataframe.loc[row_label, column_label]
Brasil = DataInmigrant.loc['Brasil', Years]
BrasilDict = {'Year':Brasil.index.tolist(),
                      'NumberMigrants':Brasil.values.tolist()}
Argentina = DataInmigrant.loc['Argentina', Years]
ArgentinaDict = {'Year':Argentina.index.tolist(),
                      'NumberMigrants':Argentina.values.tolist()}

dfBrasil = pd.DataFrame(BrasilDict)
dfArgentina = pd.DataFrame(ArgentinaDict)
```

```python
dfBrasil.head()
```

```python
dfArgentina.head()
```

```python
plt.figure(figsize=[10,5])
plt.plot(dfBrasil['Year'], dfBrasil['NumberMigrants'], color='green')
plt.plot(dfArgentina['Year'], dfArgentina['NumberMigrants'], color='blue')
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010" ])
plt.title("Argentinian-Brasilian Migration to Canda")
plt.xlabel("Year")
plt.ylabel("Number of Migrants")
plt.legend(["Brasil", "Argentina"])
plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250],
    'tiendas': ['A', 'B', 'C', 'D']}

# df = pd.DataFrame(ventas_2022, index=tiendas)
dfStores = pd.DataFrame(ventas_2022)
dfStores.set_index('tiendas', inplace = True)
dfStores.head()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Evolution of stores A, B, C, D in a Year')

months = list(dfStores.columns)
StoreA = dfStores.loc['A', months]
StoreB = dfStores.loc['B', months]
StoreC = dfStores.loc['C', months]
StoreD = dfStores.loc['D', months]

axs[0,0].plot(StoreA)
axs[0,0]. set_title('Store A')

axs[0,1].plot(StoreB)
axs[0,1]. set_title('Store B')

axs[1,0].plot(StoreC)
axs[1,0]. set_title('Store C')

axs[1,1].plot(StoreD)
axs[1,1]. set_title('Store D')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(1))
  ax.set_xlabel('Month')
  ax.set_ylabel('sales')
  ax.grid()

plt.show()
```
