# Crear Columnas Numèricas

## Crear Columna Valor_Mensual.

Esta columna debe conbtener los gastos mensuales de cada propiedad, incluyendo el alquiler y el condominio.

```python
DataDEV = pd.read_csv(url, sep=";")
DataDEV.sample(5)
```

```python
DataDEV.fillna(0, inplace=True)
DataDEV["Valor_Mensual"] = DataDEV["Valor"] + DataDEV["Condominio"]
DataDEV.head()
```

## Crear Columna valor_anual.
Esta columna debe conbtener los gastos anuales por propeidad, es decir, el impuesto mas 12 meses de alquiler y condomimio

```python
DataDEV["Valor_Anual"] = DataDEV["Valor"]*12 + DataDEV["Condominio"]*12 + DataDEV['Impuesto']
DataDEV.head()
```

## Crear Columna Descripción

Esta columna debe contener un resumen de la información clave de las propiedades que se mostrarán en el sitio web; tipo de propiedad, barrio, cantidad de habitaciones y plaza de estacionamiento

```python
DataDEV['Descripcion'] = DataDEV['Tipo'] + " - " + DataDEV['Colonia'] \
                        + " - " + DataDEV['Habitaciones'].astype(str) \
                        + " - " + DataDEV['Garages'].astype(str)
```

```python
DataDEV.head()
```

## Crear la columna Tiene_suite

esta debe ser una columna que indique úncamiente si la propiedad tiene o no suites sin la cantidad

```python
DataDEV['Tiene_Suite'] = DataDEV['Suites'].apply( lambda x: "Sí" if x > 0 else "No" )
```
