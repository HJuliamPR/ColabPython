***Demandas Machine Learning***

¿Cuáles son los valores promedio por alquiler por tipo de propiedad?

```python
Data["Valor"].mean()
```

```python
Data.head()
# Primeramente se utiliza el método groupby, luego se especifica la columna-variable que se desea intervenir por útimo se aplica el método promedio
df_priceType = Data.groupby("Tipo")[["Valor", "Impuesto"]].mean().sort_values("Valor")
df_priceType.plot(kind = "barh", figsize=(10,6), color =["purple", "red"])
```
